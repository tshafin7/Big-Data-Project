```python
from pyspark.sql.functions import col, isnan, isnull, when, count, udf
```

```python
bucket = "xxxxxxx"
filename1 = "DOT_Traffic_Speeds_NBE.csv"
filename2 = "Motor_Vehicle_Collisions_-_Crashes.csv"  


file_path1 = "gs://" + bucket + "/landing/" + filename1
file_path2 = "gs://" + bucket + "/landing/" + filename2
```

```python
sdf1 = spark.read.csv(file_path1, header=True, inferSchema=True)
sdf2 = spark.read.csv(file_path2, header=True, inferSchema=True)
```

```python
sdf1_clean = sdf1.na.drop()
```

```python
sdf2_clean = sdf2.na.drop()
```

```python
sdf1_clean = sdf1.drop("LINK_ID", "TRANSCOM Link ID", "OWNER")

```

```python
sdf2_clean = sdf2.drop(
    "LATITUDE", "LONGITUDE", "LOCATION",
    "CONTRIBUTING FACTOR VEHICLE 1", "CONTRIBUTING FACTOR VEHICLE 2", "CONTRIBUTING FACTOR VEHICLE 3",
    "CONTRIBUTING FACTOR VEHICLE 4", "CONTRIBUTING FACTOR VEHICLE 5",
    "COLLISION ID", "VEHICLE TYPE CODE 1", "VEHICLE TYPE CODE 2"
)
```

```python
output_file_path = f"gs://{bucket}/landing/cleaned"

sdf1_clean.write.mode("overwrite").parquet(f"{output_file_path}/traffic_speeds")

sdf2_clean.write.mode("overwrite").parquet(f"{output_file_path}/collisions")
```
